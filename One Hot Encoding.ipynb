```python
import pandas as pd 
```

```python
data={'Category':['A','B','C','A','B','C','A','B','C']}
```

```python
df=pd.DataFrame(data) 
df.head()
```

```python
one_hot_encoded_df=pd.get_dummies(df,columns=['Category']) 
one_hot_encoded_df
```

```python
one_hot_encoded_df=pd.get_dummies(df,columns=['Category'],prefix='Encoded',drop_first=True)
one_hot_encoded_df
```

```python
one_hot_encoded_df=pd.get_dummies(df,columns=['Category'],prefix='Dummy')
one_hot_encoded_df
```
